# QUESTION 1 : BASIC CLASS CREATION

```python
# Defining the Book class
class Book:
    def __init__(self, title, author, pages):
        # Attributes
        self.title = title
        self.author = author
        self.pages = pages

    # Method to return book info
    def get_info(self):
        return f"'{self.title}' by {self.author}, Pages: {self.pages}"

# Creating two Book objects
book1 = Book("The Power of Now", "Eckhart Tolle", 236)
book2 = Book("Atomic Habits", "James Clear", 320)

# Printing their information using the get_info() method
print(book1.get_info())
print(book2.get_info())
```

**Explanation**:

**Book Class**:

The Book class has three attributes: title, author, and pages.

**Method get_info**():

This method returns a formatted string with the book's title, author, and number of pages.

**Creating Objects**:

Two objects (book1 and book2) are created using the Book class and initialized with specific values.

**Print Information**:

The get_info() method is called on both objects to print their information.

# QUESTION 2 : CLASS INHERITANCE

```python
# Defining the Person class
class Person:
    def __init__(self, name, age):
        # Attributes of the Person class
        self.name = name
        self.age = age

# Defining the Student class that inherits from Person
class Student(Person):
    def __init__(self, name, age, student_id):
        # Inheriting attributes from Person using super()
        super().__init__(name, age)
        # Additional attribute for Student
        self.student_id = student_id

    # Method to display student details
    def display_details(self):
        return f"Student Name: {self.name}, Age: {self.age}, Student ID: {self.student_id}"

# Creating an object of the Student class
student1 = Student("SHOAN", 21, "SI303090")

# Printing the student's details
print(student1.display_details())
```

**Explanation**:


**Class Person**:

Has two attributes, name and age.

**Class Student**:

Inherits from Person and adds an additional attribute, student_id. It uses the super() function to inherit the attributes of the Person class.


**Method display_details**():

 Displays the student's details, including inherited attributes from Perso (name and age) and the student_id.

# QUESTION 3: WRITING AND READING FILES



```python
# CREATING AND WRITING TO A FILE
with open("student_data.txt", "w") as file:
    students = {"John": 85, "Emma": 90, "Sophia": 78, "Ava": 92, "Oliver": 88}
    for name, score in students.items():
        file.write(f"Name: {name}, Score: {score}\n")

# READING AND PRINTING THE CONTENTS OF THE FILE
with open("student_data.txt", "r") as file:
    for line in file:
        print(line.strip())
```

**Explanation**:

In this question, we learn how to write data to a file and read it back.

The program creates a file named student_data.txt and writes the names and scores of five students in a specified format.
It uses a for loop to iterate over a dictionary containing student names and their scores, writing each entry to the file.
After writing, it reads the file contents line by line and prints them out.

**File Writing**:

We open the file in write mode ("w"), and each student's name and score are written to the file.


**File Reading**:

 We open the file in read mode ("r"), read it line by line, and print the content.

# QUESTION 4: MANAGING STUDENT DATA WITH OOP AND FILE HANDLING

```python
# DEFINING THE STUDENTRECORD CLASS
class StudentRecord:
    def __init__(self, filename):
        self.filename = filename

    # METHOD TO ADD A STUDENT RECORD
    def add_student(self, name, score):
        with open(self.filename, "a") as file:
            file.write(f"Name: {name}, Score: {score}\n")

    # METHOD TO DISPLAY ALL RECORDS
    def display_records(self):
        with open(self.filename, "r") as file:
            for line in file:
                print(line.strip())

# CREATING AN INSTANCE AND ADDING STUDENT RECORDS
records = StudentRecord("records.txt")
records.add_student("John", 85)
records.add_student("Emma", 90)
records.add_student("Sophia", 78)

# DISPLAYING ALL RECORDS
records.display_records()
```

**Explanation**:

 This question focuses on managing student data using object-oriented programming (OOP) and file handling. We create a class StudentRecord to add and display student records.

**Attributes and Methods**:

The class has a method add_student that appends a student’s name and score to a file.
Another method, display_records, reads the file and prints the student data.

**Object Creation and Usage**:

An instance of StudentRecord is created, and student records are added. Finally, the records are displayed by reading from the file.

# QUESTION 5 : EXCEPTION HANDLING IN FILE OPERATIONS



```python
try:
    # ATTEMPTING TO OPEN A NON-EXISTENT FILE
    with open("non_existent_file.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("Error: The file does not exist. Creating the file...")

    # CREATING THE FILE AND WRITING A MESSAGE
    with open("non_existent_file.txt", "w") as file:
        file.write("This file has been created after handling the FileNotFoundError.")

    print("File has been created successfully.")
```

**Explanation**:

This question demonstrates how to handle exceptions in file operations, specifically dealing with the FileNotFoundError.

**File Opening and Error Handling**:

The program tries to open a file that doesn’t exist (non_existent_file.txt).

When the file isn’t found, the FileNotFoundError exception is caught, and an error message is displayed.

**File Creation**:

After handling the exception, the file is created, and a custom message is written to it. The program then confirms that the file has been created.
